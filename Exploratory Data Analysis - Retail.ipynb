# Profit Analysis

# Analysis of Profit, Sales and Discount

## Segment Wise Analysis 

```python
df['Segment'].value_counts()
```

```python
df_segment= df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```

```python
df_segment.plot.pie(subplots = True,
                    autopct='%1.1f%%',
                    figsize=(20,18),
                    labels = df_segment.index)
plt.title('Segment Wise Analysis of Sales, Discount, Profit', fontsize='20')
```

### From the above plots, we can infer that
    For Sales, Home Office has the highest number of sales
    For Discount, Customer and Corporate has highest number of discount
    For Profit, Home Office has the highest number of profit

## Ship Mode Wise Analysis 

```python
df['Ship Mode'].value_counts()
```

```python
df_shipmode= df.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean()
df_shipmode
```

```python
df_shipmode.plot.pie(subplots = True,
                    autopct='%1.1f%%',
                    figsize=(20,18),
                    labels = df_shipmode.index)
plt.title('Ship Mode Wise Analysis of Sales, Discount, Profit', fontsize='20')
```

### From the above plots, we can infer that
     For Sales, Same Day has the highest number of sales
     For Discount, First Class has the highest number of discount
     For Profit, First Class has the highest number of profit
